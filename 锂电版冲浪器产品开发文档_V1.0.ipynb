# 锂电版冲浪器产品开发文档 V1.0

编制日期：2025-10-09    项目负责人：张宏

## 一、项目概述
产品定义：挂装于泳池岸边台阶的锂电驱动水流跑步机，主机浸入水体工作，经叶轮（卡普兰型）与导流筒生成稳定逆向水流，使用户在原地游泳开展持续训练与娱乐。

## 二、产品定位与使用场景  
核心用户：高端家庭泳池业主、别墅/公寓会所、泳池私教工作室、康复理疗机构、精品度假酒店。  
典型场景：1）家庭体能训练与儿童趣味游玩；2）私教课程中的定点抵抗泳训；3）康复中心低冲击有氧训练；4）酒店或度假村水疗配套体验。  
价值主张：无需改造泳池即可部署；占地小、噪声低；提供稳定宽幅水流，兼顾专业训练与娱乐；模块化设计便于维护和租赁运营。

## 四、主要技术参数
尺寸与重量：主机720×320×285mm；电源箱380×220×180mm；整机含支架26kg（主机15kg、电源箱11kg）。  
水流性能：喷口截面170×109mm，五档出口流速 0.6/1.0/1.8/2.6/3.6m/s，对应流量 180–230m³/h，核心区流速均匀性 ±8% 以内。  
动力系统：额定功率 1.0kW、峰值 1.4kW 无刷直流电机，最高转速 1900rpm，效率 ≥88%，采用水润滑轴承与双端机械密封。  
能源系统：24V/62.5Ah（1.5kWh）锂电模组；典型3档续航 70min，5 档持续 45min；配 700W 充电器，满充 4h。  
声学与振动：水下噪声 ≤58dB(A)@1m；挂架振动加速度 ≤1.2m/s²；喷口脉动压力 ≤8kPa。  
环境适应：工作水温 10–45℃；主机防护 IP69K、电源箱 IP66；盐雾 120h 无结构失效。

## 五、结构与材料设计
外壳与框架：ASA原色注塑外壳；2205双相不锈钢钣金；  
流道与喷口：PPO原色注塑粘接流道，双层导流板，90度，圆口转宽方口；  
进水防护：进水口附近钣金、外壳开孔宽度小于8mm满足手指安规测试要求，内部流道入口加装3mm小孔过滤罩，防止头发卷入叶轮；  
密封体系：电机上部机械动密封，下部双层O型圈静密封；水冷驱动器整体灌胶密封；动力线插头胶塞密封；电源箱外壳密封圈密封；按钮O型圈密封；插头密封垫密封；胶塞密封；  
高度调节机构：316L可调支架；橡胶防滑垫；  
减振设计：电机EPDM减振安装座；流道筒出入口刚性圈；大圆弧分层流道设计；

## 六、流体力学设计
流道规划：外壳外水流穿过外壳前后左右上的孔，进入内部，穿过流道入口处的圆孔过滤筒，进入流道入口处，叶轮驱动水流向上运动，经过整流导轮，进入90度大半径弯流道，经导流板分层导流，在出水隔栅处分层喷出。  

## 七、电气系统设计  
电池模组：13s12p 21700 高倍率电芯，BMS 提供 SOH/SOC 估算、主动均衡、过温/过流/短路/绝缘监测，支持 CAN 与 BLE 通讯。  
主控单元：Cortex-M33 MCU 负责档位管理、闭环转速控制、温度和振动监测，并记录运行日志。  
功率驱动：三相无刷 FOC 控制器，MOSFET 模块额定 150A，配水冷板与导热硅胶垫，通过工作水体散热。  
人机界面：控制箱触感按键与 LED 显示档位、电量、故障码；腕绳磁控开关 200ms 内急停；App 支持档位曲线设定、维护提醒与固件更新。  
安全逻辑：电池舱开盖、温度≥70℃、漏电或连接器未锁止时禁止启机；急停后需复位并重新上电。

## 八、制造工艺与装配  
塑件工艺：原色注塑，外观件晒纹处理，抗UV老化；  
金属加工：2205双相不锈钢钣金电解钝化，316不锈钢电解钝化，金属标准件点解钝化；  
装配流程：前后壳隔栅组装→框架安装→滚轮支撑安装→电机、过滤筒、叶轮与驱动模块安装→流道装配→叶轮间隙检测→外壳安装→整机气密/耐压测试→控制箱线束装配→老化与功能验证。  
工装夹具：主机定位夹具、O 形圈装配治具、动力线压接台、喷口流速测试水槽、控制箱防水检具。  
质量控制：关键螺栓扭矩记录，IP69K 喷淋抽检，动力线绝缘电阻 ≥50MΩ，充放电循环测试 ≥300 次。

## 九、性能测试与验证

水流测试：在 10×5m 标准水池布设 3×3 测点，使用声学多普勒流速仪测定各档流速分布和衰减。

续航与热管理：典型 3 档/5 档工况连续 90min 测温，记录电机绕组、驱动器、电池模组温升，确认处于安全范围。

防护验证：执行 IP69K 喷淋、2m 静态浸泡 1h、动力线弯折 10,000 次；控制箱完成 IP66 喷淋与 1m 跌落测试。

安全测试：腕绳断开响应 ≤0.2s；急停后 3s 内流速降至 0.4m/s 以下；漏电保护动作电流 15mA。

用户评测：组织私教、康复师、家庭用户各 10 组体验，收集舒适度、噪声、安装便捷性反馈并形成改进清单。

## 十、可靠性与安全性设计

结构可靠性：主机壳体通过 0.3MPa 静压和 5J 冲击测试；挂架静载 1.5 倍、循环 50,000 次无失效。

电气安全：符合 IEC 60335-2-60、GB 4706.1，具备漏电检测与绝缘监测，欠压/过压/过温自动降档或停机。

热管理：电机、驱动器、电池多点温度监测，70℃ 分级降功率，80℃ 强制停机；控制箱采用风道与相变散热模块。

人机安全：全覆盖进水格栅防夹手；喷口防护网可拆洗；动力线防误拔锁扣；儿童安全模式限制最高档位。

可靠性指标：目标 MTBF ≥2500h，B10 寿命 1200h；电机、密封、驱动等关键部件完成 1500h 连续寿命试验。

## 十一、成本估算与可制造性分析

原型成本：主机结构与流道 3200 元，电机与驱动 3600 元，锂电模组 5200 元，控制箱与线束 1800 元，装配与测试 1600 元。

量产策略：塑件模具摊销后成本下降 30%；驱动器与电机年度采购降 15%；线束标准化减少装配工时 25%。

生产节拍：两人一线目标 40min/套，引入半自动点胶与智能拧紧提高一致性。

供应链：电机、锂电模组、轴封等关键件建立双供应，国产化率预计 75%，核心物料交期 6 周。

## 十二、未来优化方向

智能化：开发水流自适应算法与心率/动作识别联动，实现个性化训练计划并接入云端平台。

产品矩阵：规划超宽喷口高流量版、轻量便携电源包、联网租赁版以及可定制外观面板。

声学优化：研究螺旋桨尾迹噪声与结构传递路径，采用多层阻尼材料和主动降噪策略。

安装便捷：开发免工具快装支架、吸盘+膨胀螺栓组合方案，适配无台阶或不规则岸边。

可持续设计：建立电池回收与梯次利用计划，引入可替换饰板与可循环包装，通过绿色供应链认证。
